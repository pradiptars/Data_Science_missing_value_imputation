Data Clining missing value inputing


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.head(6)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.info()
```

```python
missing_value_per = df.isnull().sum()/df.shape[0]*100
missing_value_per
```

```python
missing_value_clm = missing_value_per[missing_value_per>4].keys()
missing_value_clm
```

```python
df3_drop_clm = df.drop(columns=missing_value_clm)
df3_drop_clm.shape
```

```python
df.shape
```

```python
df3_num = df3_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

Finding Missing data

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(25,25))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color': '#DC143C'})
```

Filling Data

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
```

```python
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth': 8, 'color': 'green'}, label= 'Original')
  sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth': 5, 'color': 'red'}, label= 'Mean')
  plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df3_num[var], bins=20, hist=False, kde_kws={'linewidth': 8, 'color': 'green'}, label= 'Original')
  sns.distplot(df4_num_mean[var], bins=20, hist=False, kde_kws={'linewidth': 5, 'color': 'red'}, label= 'Mean')
  sns.distplot(df5_num_median[var], bins=20, hist=False, kde_kws={'linewidth': 3, 'color': 'blue'}, label= 'Median')
  plt.legend()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df[var], bins=20, hist=False, kde_kws={'linewidth': 8, 'color': 'green'}, label= 'Original')
  sns.distplot(df4_num_mean[var], bins=20, hist=False, kde_kws={'linewidth': 5, 'color': 'red'}, label= 'Mean')
  sns.distplot(df5_num_median[var], bins=20, hist=False, kde_kws={'linewidth': 3, 'color': 'blue'}, label= 'Median')
  plt.legend()
```

```python
for i, var in enumerate(missing_num_var):
  plt.figure(figsize=(10,10))
  plt.subplot(3,1,1)
  sns.boxplot(df[var])
  plt.subplot(3,1,2)
  sns.boxplot(df4_num_mean[var])
  plt.subplot(3,1,3)
  sns.boxplot(df5_num_median[var])
```

```python
df_concat = pd.concat([df3_num[missing_num_var], df4_num_mean[missing_num_var], df5_num_median[missing_num_var]], axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
